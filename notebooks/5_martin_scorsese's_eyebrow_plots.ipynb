```python
def mymap(xthing,ything,zthing,xlabel,ylabel,x1,x2,y1,y2):
    t = binned_statistic_2d(xthing,ything,zthing,statistic = 'count',bins=(100,100))
    yedges = t[1]
    xedges = t[2]
    number = np.ma.array(t[0], mask = t[0] == 0)
    number = np.log10(number)
    #plt.plot([x1,x2],[y1,y2], color = 'grey', linestyle = '--')
    plt.pcolormesh(yedges,xedges,(np.transpose(number)),cmap = plt.cm.viridis_r)
    plt.colorbar()
    plt.xlabel(str(xlabel))
    plt.ylabel(str(ylabel))
    plt.xlim(x1,x2)
    plt.ylim(y1,y2)
    plt.show()
```
