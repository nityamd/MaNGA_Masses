```python
print(len(df))
df = df[df['sfr_tot_p50']>= -100]
```

```python
df = df[df['d4000_n']>=0][df['d4000_n']<=2.3]
print(len(df))
```

```python

fig, ((ax1, ax2), (ax3,ax4),(ax5,ax6)) = plt.subplots(3, 2, figsize = (10,14), sharex = 'col', sharey = 'row')
#fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (13,7), sharey = True)

a = binned_statistic(df['d4000_n'], df['sfr_tot_p50'], statistic = 'mean',bins =20)
xbins_a = [0.5*(a[1][i] + a[1][i+1]) for i in range(len(a[1])-1)]
b = binned_statistic(df['lick_hd_a'], df['sfr_tot_p50'], statistic = 'mean',bins =20)
xbins_b = [0.5*(b[1][i] + b[1][i+1]) for i in range(len(b[1])-1)]

ax1.plot(xbins_a,a[0])
ax1.set_xlim(0.8,2.2)
ax1.set_ylabel('Mean SFR')
ax2.plot(xbins_b,b[0])

a = binned_statistic(df['d4000_n'], df['lgm_tot_p50'], statistic = 'mean',bins =20)
xbins_a = [0.5*(a[1][i] + a[1][i+1]) for i in range(len(a[1])-1)]
b = binned_statistic(df['lick_hd_a'], df['lgm_tot_p50'], statistic = 'mean',bins =20)
xbins_b = [0.5*(b[1][i] + b[1][i+1]) for i in range(len(b[1])-1)]

ax3.plot(xbins_a,a[0])
ax3.set_ylabel('Stellar Mass')
ax4.plot(xbins_b,b[0])

a = binned_statistic(df['d4000_n'], df['mlratio_z'], statistic = 'mean',bins =20)
xbins_a = [0.5*(a[1][i] + a[1][i+1]) for i in range(len(a[1])-1)]
b = binned_statistic(df['lick_hd_a'], df['mlratio_z'], statistic = 'mean',bins =20)
xbins_b = [0.5*(b[1][i] + b[1][i+1]) for i in range(len(b[1])-1)]
print(len(a[0]), len(a[1]))


ax5.plot(xbins_a,a[0])
ax5.set_xlim(0.8,2.2)
ax5.set_xlabel(r'$D_{\rm n}4000$')
ax5.set_ylabel(r'${\rm log}_{10}(\rm M/L)_{z}$')
ax6.plot(xbins_b,b[0])
ax6.set_xlim(-10,10)
ax6.set_xlabel(r'$H\delta_{\rm A}$')

plt.gcf().subplots_adjust(wspace=0, hspace = 0, bottom =  0.13)
plt.savefig("/Users/Nitya/Desktop/mass_age_sfr_dist.pdf")
plt.show()
```
